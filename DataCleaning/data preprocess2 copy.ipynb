```python
length=len(mpg)
d=dict()
#for i in range(length):
for i in range(length):
    row=mpg.iloc[i,]
    userid=row['user id']
    movieid=row['movie id']
    timing=row['timing']
    if (userid,movieid) not in d:
        d[(userid,movieid)]=timing
    else:
        if timing>d[(userid,movieid)]:
            d[(userid,movieid)]=timing

    
```

```python
df = pd.DataFrame(columns=['time','user id','movie id','year','rating'])
count=0
length=len(mpg)
for i in range(length):
    row=mpg.iloc[i,]
    userid=row['user id']
    movieid=row['movie id']
    timing=row['timing']
    if d[(userid,movieid)]==timing:
        val=0
        if timing<=20:
            val=1
        elif timing<=50:
            val=2
        elif timing<=100:
            val=3
        elif timing<=150:
            val=4
        elif timing>150:
            val=5
        print("Adding row ",i)
        df.loc[count,]=row
        df.iloc[count,]['rating']=val
        count+=1
df.head()
```
